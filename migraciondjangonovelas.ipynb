```python
import pandas as pd
import os
import requests
import json
import time
```

```python
# path = os.path.join(os.getcwd(),'chapters')
nombrenovel = 'I_Reincarnated_As_An_Immortal_Bat_Into_Another_World.csv'
novelchapters = os.path.join(os.getcwd(), 'chapters', nombrenovel)
novelcontentchapters = os.path.join(os.getcwd(), 'completes', nombrenovel)
novela_id = '65f323880482fb051520953e'
# os.listdir(path)
print(novelchapters)
dfchapters = pd.read_csv(novelchapters)
if "novela_id" not in dfchapters.columns:
    dfchapters["novela_id"]=novela_id
if "capitulo_id" not in dfchapters.columns:
    dfchapters["capitulo_id"]=""
dfcontentchapters = pd.read_csv(novelcontentchapters)
print(dfcontentchapters.columns)
dfchapters
```

```python
urlcapitulos = "http://localhost:8000/api/capitulos/"
# urlscontenidocaps = "http://localhost:8000/api/contenidocapitulos/"
headers = {
  'Content-Type': 'application/json'
}
```

```python
idx = 0
while idx < len(dfchapters):
    payloadchapter = json.dumps({
        "novela_id": str(dfchapters['novela_id'][idx]),
        "nombre": str(dfchapters['nombre'][idx]),
        "url": str(dfchapters['urls'][idx])
    })
    rescap = requests.request("POST", urlcapitulos, headers=headers, data=payloadchapter)
    idcap = json.loads(str(rescap.text))['_id']
    print(idcap)
    dfchapters['capitulo_id'][idx]=idcap
    idx = idx + 1
    # break
```

```python
dfchapters.to_csv(novelchapters)
dfchapters
```
